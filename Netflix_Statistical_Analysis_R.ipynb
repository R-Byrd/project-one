```python
import pandas as pd
import numpy as np
import json
from pathlib import Path
import matplotlib.pyplot as plt
import os


netflix_data = Path("Resources/updated_cleaned_ended_netflix_shows_05_15_V3.csv")
```

```python
#Reading csv file
netflix_data_df = pd.read_csv(netflix_data, encoding="ISO-8859-1")
netflix_data_df.columns
```

```python
netflix_data_df.describe()
```

```python
# Count the number of TV shows per year
shows_per_year = netflix_data_df['Premiere Year'].value_counts().sort_index()

# Print the results
print(shows_per_year)

# Plot the number of TV shows per year
plt.figure(figsize=(10, 6))
shows_per_year.plot(kind='bar')
plt.title('Number of TV Shows per Year')
plt.xlabel('Year of Premiere')
plt.ylabel('Number of TV Shows')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

```

```python
# Filter the DataFrame to include only canceled shows (Season Canceled == 1)
canceled_shows_df = netflix_data_df.loc[netflix_data_df['Seasons'] == 1 ]

# Count the number of canceled shows per finale year
canceled_shows_per_year = canceled_shows_df['Finale Year'].value_counts().sort_index()

# Print the results
print(canceled_shows_per_year)

# Plot the number of canceled shows per finale year
plt.figure(figsize=(10, 6))
ax = canceled_shows_per_year.plot(kind='bar')
plt.title('Number of Canceled TV Shows per Finale Year')
plt.xlabel('Year')
plt.ylabel('Number of Canceled TV Shows')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Format the x-axis labels to remove the decimal
ax.set_xticklabels([int(label.get_text().replace('.0', '')) for label in ax.get_xticklabels()])

# Display the plot
plt.show()
```
