```python
import pandas as pd
import numpy as np
import json
from pathlib import Path
import requests

#Defining file paths
netflix_titles = Path("Resources/netflix_titles.csv")
ended_netflix_shows = Path("Resources/All Ended Netflix Original Shows.csv")
netflix_data = Path("Resources/netflix_data.csv")

print(netflix_titles)
```

```python
#Reading data files
netflix_titles_df = pd.read_csv(netflix_titles, encoding="ISO-8859-1")
ended_netflix_shows_df = pd.read_csv(ended_netflix_shows, encoding="ISO-8859-1")
netflix_data_df = pd.read_csv(netflix_data, encoding="ISO-8859-1")

netflix_data_df.head()
```

```python
ended_netflix_shows_df.rename(columns={'Title': 'title'}, inplace=True)
netflix_titles_df.rename(columns={'listed_in': 'Genre'}, inplace=True)
netflix_data_df.columns
```

```python
netflix_merge = pd.merge(ended_netflix_shows_df, netflix_titles_df, on='title', how='outer')
netflix_combined_df = pd.merge(netflix_merge, netflix_data_df, on='title', how='outer')

netflix_combined_df.columns
```

```python
# Drop Unnamed and duplicates

netflix_combined_df.drop(['Genre_x', 'Genre_y', 'show_id_y', 'show_id_x', 'duration_y', 'description_y', 'description_x', 'show_id_x', 'show_id_y', 'director_y', 'cast_y', 'director_x', 'cast_x','Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',
                          'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22',
                          'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25'], axis=1, inplace=True)



netflix_combined_df.head()
```

```python
# Function to combine country_x and country_y into a new column Country without duplicates
def combine_countries(row):
    if pd.notna(row['country_x']) and pd.notna(row['country_y']):
        if row['country_x'] == row['country_y']:
            return row['country_x']
        else:
            return f"{row['country_x']}, {row['country_y']}"
    elif pd.notna(row['country_x']):
        return row['country_x']
    else:
        return row['country_y']

# Apply the function to create the new Country column
netflix_combined_df['Country'] = netflix_combined_df.apply(combine_countries, axis=1)

# Drop the original country_x and country_y columns
netflix_combined_df = netflix_combined_df.drop(columns=['country_x', 'country_y'])

# Display the updated DataFrame
print(netflix_combined_df.head())
```

```python
# Function to combine type_x and type_y into a new column Type without duplicates
def combine_countries(row):
    if pd.notna(row['type_x']) and pd.notna(row['type_y']):
        if row['type_x'] == row['type_y']:
            return row['type_x']
        else:
            return f"{row['type_x']}, {row['type_y']}"
    elif pd.notna(row['type_x']):
        return row['type_x']
    else:
        return row['type_y']

# Apply the function to create the new Type column
netflix_combined_df['Type'] = netflix_combined_df.apply(combine_countries, axis=1)

# Drop the original type_x and type_y columns
netflix_combined_df = netflix_combined_df.drop(columns=['type_x', 'type_y'])

# Display the updated DataFrame
print(netflix_combined_df.head())
```
