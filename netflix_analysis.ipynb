```python
import pandas as pd
import numpy as np
import json
from pathlib import Path
import requests

#Defining file paths
netflix_titles = Path("Resources/netflix_titles.csv")
ended_netflix_shows = Path("Resources/All Ended Netflix Original Shows.csv")
netflix_data = Path("Resources/netflix_data.csv")

print(netflix_titles)
```

```python
netflix_merge = pd.merge(ended_netflix_shows_df, netflix_titles_df, on='title', how='outer')
netflix_combined_df = pd.merge(netflix_merge, netflix_data_df, on='title', how='outer')

netflix_combined_df.columns
```

```python
specific_title = "Orange Is the New Black"

# Filter the DataFrame to get rows where the 'title' column matches the specific title
filtered_rows = netflix_combined_df[netflix_combined_df['title'] == specific_title]

# Display the filtered rows
print(filtered_rows)
```

```python
def combine_genres(row):
    if pd.isna(row['Genre_x']) and pd.isna(row['Genre_y']):
        return None  # or a default string like 'Unknown' if you prefer
    if pd.isna(row['Genre_x']):
        return row['Genre_y']
    if pd.isna(row['Genre_y']):
        return row['Genre_x']
    return row['Genre_x'] + ' & ' + row['Genre_y']

# Apply the function to each row to create the Genre column
netflix_combined_df['Genre'] = netflix_combined_df.apply(combine_genres, axis=1)

# Display the updated DataFrame to verify the new column
netflix_combined_df.head()
```

```python
def combine_types(row):
    if pd.isna(row['type_x']) and pd.isna(row['type_y']):
        return None  # or a default string like 'Unknown' if you prefer
    if pd.isna(row['type_x']):
        return row['type_y']
    if pd.isna(row['type_y']):
        return row['type_x']
    return row['type_x'] + ' & ' + row['type_y']

# Apply the function to each row to create the Type column
netflix_combined_df['Type'] = netflix_combined_df.apply(combine_types, axis=1)

# Display the updated DataFrame to verify the new column
netflix_combined_df.head()
```

```python
netflix_combined_df.drop(['Genre_x', 'type_x', 'type_y',
       'Genre_y'], axis=1, inplace=True)
```

```python
def drama_genre(genre):
    if genre is None:  # Check if the genre is None and return it as is if true
        return genre
    if 'drama' in genre.lower():  # Case-insensitive check for 'drama'
        return 'Dramas'
    else:
        return genre
# Apply the function to the 'Genre' column
netflix_combined_df['Genre'] = netflix_combined_df['Genre'].apply(drama_genre)
```
