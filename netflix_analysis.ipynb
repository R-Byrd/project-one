```python
import pandas as pd
import numpy as np
import json
from pathlib import Path
import requests

#Defining file paths
netflix_titles = Path("Resources/netflix_titles.csv")
ended_netflix_shows = Path("Resources/All Ended Netflix Original Shows.csv")

print(netflix_titles)
```

```python
#Reading data files
netflix_titles_df = pd.read_csv(netflix_titles, encoding="ISO-8859-1")
ended_netflix_shows_df = pd.read_csv(ended_netflix_shows, encoding="ISO-8859-1")

ended_netflix_shows_df.head()
```

```python
ended_netflix_shows_df.rename(columns={'Title': 'title'}, inplace=True)
netflix_titles_df.rename(columns={'listed_in': 'Genre'}, inplace=True)
ended_netflix_shows_df.columns
```

```python
netflix_combined_df = pd.merge(ended_netflix_shows_df, netflix_titles_df, on='title', how='outer')
num_rows = len(netflix_combined_df)
netflix_combined_df.columns
```

```python
netflix_combined_df.drop(['description', 'show_id', 'director', 'cast','Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',
                          'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22',
                          'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25'], axis=1, inplace=True)

```

```python
def combine_genres(row):
    if pd.isna(row['Genre_x']) and pd.isna(row['Genre_y']):
        return None  # or a default string like 'Unknown' if you prefer
    if pd.isna(row['Genre_x']):
        return row['Genre_y']
    if pd.isna(row['Genre_y']):
        return row['Genre_x']
    return row['Genre_x'] + ' & ' + row['Genre_y']

# Apply the function to each row to create the Genre column
netflix_combined_df['Genre'] = netflix_combined_df.apply(combine_genres, axis=1)

# Display the updated DataFrame to verify the new column
netflix_combined_df.head()
```

```python
def combine_types(row):
    if pd.isna(row['type_x']) and pd.isna(row['type_y']):
        return None  # or a default string like 'Unknown' if you prefer
    if pd.isna(row['type_x']):
        return row['type_y']
    if pd.isna(row['type_y']):
        return row['type_x']
    return row['type_x'] + ' & ' + row['type_y']

# Apply the function to each row to create the Type column
netflix_combined_df['Type'] = netflix_combined_df.apply(combine_types, axis=1)

# Display the updated DataFrame to verify the new column
netflix_combined_df.head()
```

```python
genres = netflix_combined_df["Genre_x"].unique()

len(genres)
```

```python
def replace_genre(genre):
    # Check if the genre is a string and contains 'Documentaries'
    if pd.notna(genre) and 'Drama' in genre:
        return 'Dramas'
    else:
        return genre

# Check if 'genre_x' and 'genre_y' exist in the DataFrame
if 'Genre_x' in netflix_combined_df.columns and 'Genre_y' in netflix_combined_df.columns:
    # Apply the function to 'genre_x' and 'genre_y' columns
    netflix_combined_df['Genre_x'] = netflix_combined_df['Genre_x'].apply(replace_genre)
    netflix_combined_df['Genre_y'] = netflix_combined_df['Genre_y'].apply(replace_genre)
    # Display the updated DataFrame to verify changes
    print(netflix_combined_df.head())
else:
    print("Check the column names. 'genre_x' or 'genre_y' not found in the DataFrame.")
```
